```python
import seaborn as sns
import pandas as pd
import os
import sys
from dotenv import find_dotenv, load_dotenv
load_dotenv(find_dotenv())
```

```python
exam_answers_path = os.getenv("EXAM_ANSWERS")
```

```python
scores = pd.read_csv(os.path.join(exam_answers_path, "classroom_data.csv"))
```

```python
scores = scores[scores['top_k']==5]
```

```python
sns.lineplot(data=scores, x="chunk_size", y="mean_score")
sns.scatterplot(data=scores, x="chunk_size", y="mean_score")
```
